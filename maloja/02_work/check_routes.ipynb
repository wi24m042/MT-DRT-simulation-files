```python
for routes_filepath in routes_filepath_list:

    routes_file_xmltree = ET.parse(routes_filepath)
    routes_file_xmlroot = routes_file_xmltree.getroot()

    print(routes_filepath)

    pc = 0
    pptc = 0
    pptr = 0

    for person in list(routes_file_xmlroot):
        pc += 1
        for leg in person:
            if leg.tag == "ride":
                pptr+=1
                if "intended" in leg.attrib:
                    pptc += 1
                    break
    print("person count: " + str(pc) + " persons | persons with rides: " + str(pptr) + " | persons with public transportation: " + str(pptc))
```
