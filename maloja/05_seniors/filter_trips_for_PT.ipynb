```python
import xml.etree.ElementTree as ET
import copy
```

```python
persontrips_filepath = "D:\\Dokumente\\sumo\\maloja\\05_seniors\\persontrips_maloja_seniors.xml"
persontrips_filepath_output = "D:\\Dokumente\\sumo\\maloja\\05_seniors\\persontrips_maloja_seniors_modded.xml"
routes_filepath = "D:\\Dokumente\\sumo\\maloja\\05_seniors\\persontrips_maloja_seniors_resolved_offseason_bus.rou.xml"
routes_filepath_output = "D:\\Dokumente\\sumo\\maloja\\05_seniors\\persontrips_maloja_seniors_resolved_offseason_bus_modded.rou.xml"
```

```python
demand = {
    "taz_Bregaglia-taz_Bregaglia": 23,
    "taz_Sils_Engadin-taz_Sils_Engadin": 1,
    "taz_Silvaplana-taz_Silvaplana": 2,
    "taz_St_Moritz-taz_St_Moritz": 55,

    "taz_Bregaglia-taz_Sils_Engadin": 23,
    "taz_Bregaglia-taz_Silvaplana": 23,
    "taz_Bregaglia-taz_St_Moritz": 23,
    "taz_Bregaglia-taz_St_Moritz_Sink": 23,

    "taz_Sils_Engadin-taz_Bregaglia": 7,
    "taz_Sils_Engadin-taz_Silvaplana": 7,
    "taz_Sils_Engadin-taz_St_Moritz": 7,
    "taz_Sils_Engadin-taz_St_Moritz_Sink": 7,

    "taz_Silvaplana-taz_Bregaglia": 15,
    "taz_Silvaplana-taz_Sils_Engadin": 15,
    "taz_Silvaplana-taz_St_Moritz": 15,
    "taz_Silvaplana-taz_St_Moritz_Sink": 15,

    "taz_St_Moritz-taz_Bregaglia": 55,
    "taz_St_Moritz-taz_Sils_Engadin": 55,
    "taz_St_Moritz-taz_Silvaplana": 55,
    "taz_St_Moritz-taz_St_Moritz_Sink": 55
}
```

```python
demand_leaving = copy.deepcopy(demand)
demand_returning = copy.deepcopy(demand)
```

```python
persontrips_file_xmltree = ET.parse(persontrips_filepath)
persontrips_file_xmlroot = persontrips_file_xmltree.getroot()
```

```python
routes_file_xmltree = ET.parse(routes_filepath)
routes_file_xmlroot = routes_file_xmltree.getroot()
```

```python
for trip_child in list(routes_file_xmlroot):
    has_ride = False
    for stage_child in list(trip_child):
        if stage_child.tag == "ride":
            if "intended" in stage_child.attrib:
                has_ride = True
                break

    if has_ride == True:
        for persontrip_child in list(persontrips_file_xmlroot):
            if trip_child.get("id") == persontrip_child.get("id"):
                demand_key_leaving = str(persontrip_child[0].get("fromTaz")) + "-" + str(persontrip_child[0].get("toTaz"))
                demand_key_returning = str(persontrip_child[0].get("toTaz")) + "-" + str(persontrip_child[0].get("fromTaz"))
                
                if 28800 <= float(str(trip_child.get("depart"))) <= 50400 and demand_key_leaving in demand_leaving and demand_leaving[demand_key_leaving] > 0:
                    demand_leaving[demand_key_leaving] = demand_leaving[demand_key_leaving] - 1
                    break
                elif 39600 <= float(str(trip_child.get("depart"))) <= 64800 and demand_key_returning in demand_returning and demand_returning[demand_key_returning] > 0:
                    demand_returning[demand_key_returning] = demand_returning[demand_key_returning] - 1
                    break
                else:
                    routes_file_xmlroot.remove(trip_child)
                    break

    if has_ride == False:
        routes_file_xmlroot.remove(trip_child)
```

```python
for persontrip_child in list(persontrips_file_xmlroot):
    persontrip_in_routes = False
    for trip_child in list(routes_file_xmlroot):
        if trip_child.get("id") == persontrip_child.get("id"):
            persontrip_in_routes = True
            break
    
    if persontrip_in_routes == False:
        persontrips_file_xmlroot.remove(persontrip_child)
```

```python
persontrips_file_xmltree.write(persontrips_filepath_output)
routes_file_xmltree.write(routes_filepath_output)
```
