```python
import xml.etree.ElementTree as ET
import copy
```

```python
persontrips_filepath = "D:\\Dokumente\\sumo\\maloja\\04_shopping\\persontrips_maloja_shopping_modded.xml"
persontrips_filepath_output = "D:\\Dokumente\\sumo\\maloja\\04_shopping\\persontrips_maloja_shopping_modded_doubled.xml"
routes_filepath_output = "D:\\Dokumente\\sumo\\maloja\\04_shopping\\persontrips_maloja_shopping_resolved_offseason_bus_modded.rou.xml"
routes_filepath_output = "D:\\Dokumente\\sumo\\maloja\\04_shopping\\persontrips_maloja_shopping_resolved_offseason_bus_modded_doubled.rou.xml"
```

```python
persontrips_file_xmltree = ET.parse(persontrips_filepath)
persontrips_file_xmlroot = persontrips_file_xmltree.getroot()
```

```python
routes_file_xmltree = ET.parse(routes_filepath_output)
routes_file_xmlroot = routes_file_xmltree.getroot()
```

```python
for person in list(persontrips_file_xmlroot):
    if "gb" not in str(person.get("id")) or "ga" not in str(person.get("id")):
        person2 = copy.deepcopy(person)
        person2.set("id", "ga" + str(person2.get("id")))
        persontrips_file_xmlroot.insert(list(persontrips_file_xmlroot).index(person), person2)
        
        person.set("id", "gb" + str(person.get("id")))
```

```python
for person in list(routes_file_xmlroot):
    if "gb" not in str(person.get("id")) or "ga" not in str(person.get("id")):
        person2 = copy.deepcopy(person)
        person2.set("id", "ga" + str(person2.get("id")))
        routes_file_xmlroot.insert(list(routes_file_xmlroot).index(person), person2)
        
        person.set("id", "gb" + str(person.get("id")))
```

```python
persontrips_file_xmltree.write(persontrips_filepath_output)
routes_file_xmltree.write(routes_filepath_output)
```
