```python
import xml.etree.ElementTree as ET
```

```python
persontrips_filepath = "D:\\Dokumente\\sumo\\maloja\\04_shopping\\persontrips_maloja_shopping.xml"
persontrips_filepath_output = "D:\\Dokumente\\sumo\\maloja\\04_shopping\\persontrips_maloja_shopping_modded.xml"
routes_filepath = "D:\\Dokumente\\sumo\\maloja\\04_shopping\\persontrips_maloja_shopping_resolved_offseason_bus.rou.xml"
routes_filepath_output = "D:\\Dokumente\\sumo\\maloja\\04_shopping\\persontrips_maloja_shopping_resolved_offseason_bus_modded.rou.xml"
```

```python
demand = {
    "taz_Bregaglia-taz_shopping_st_moritz_west": 3,
    "taz_Bregaglia-taz_shopping_bregaglia_vicosoprano": 5,
    "taz_Bregaglia-taz_shopping_bregaglia_maloja": 2,
    "taz_Bregaglia-taz_shopping_bregaglia_castasegna": 2,
    "taz_Bregaglia-taz_shopping_bregaglia_bondo": 2,

    "taz_Sils_Engadin-taz_shopping_st_moritz_west": 3,
    "taz_Sils_Engadin-taz_shopping_sils_engadin": 3,

    "taz_Silvaplana-taz_shopping_st_moritz_west": 5,
    "taz_Silvaplana-taz_shopping_silvaplana": 5,

    "taz_St_Moritz-taz_shopping_st_moritz_west": 34,
    "taz_St_Moritz-taz_shopping_st_moritz_center": 15,


    "taz_shopping_st_moritz_west-taz_Bregaglia": 3,
    "taz_shopping_bregaglia_vicosoprano-taz_Bregaglia": 5,
    "taz_shopping_bregaglia_maloja-taz_Bregaglia": 2,
    "taz_shopping_bregaglia_castasegna-taz_Bregaglia": 2,
    "taz_shopping_bregaglia_bondo-taz_Bregaglia": 2,

    "taz_shopping_st_moritz_west-taz_Sils_Engadin": 3,
    "taz_shopping_sils_engadin-taz_Sils_Engadin": 3,

    "taz_shopping_st_moritz_west-taz_Silvaplana": 5,
    "taz_shopping_silvaplana-taz_Silvaplana": 5,

    "taz_shopping_st_moritz_west-taz_St_Moritz": 34,
    "taz_shopping_st_moritz_center-taz_St_Moritz": 15
}
```

```python
persontrips_file_xmltree = ET.parse(persontrips_filepath)
persontrips_file_xmlroot = persontrips_file_xmltree.getroot()
```

```python
routes_file_xmltree = ET.parse(routes_filepath)
routes_file_xmlroot = routes_file_xmltree.getroot()
```

```python
for trip_child in list(routes_file_xmlroot):
    has_ride = False
    for stage_child in list(trip_child):
        if stage_child.tag == "ride":
            if "intended" in stage_child.attrib:
                has_ride = True
                break

    if has_ride == True:
        for persontrip_child in list(persontrips_file_xmlroot):
            if trip_child.get("id") == persontrip_child.get("id"):
                demand_key = str(persontrip_child[0].get("fromTaz")) + "-" + str(persontrip_child[0].get("toTaz"))
                if demand[demand_key] > 0:
                    demand[demand_key] = demand[demand_key] - 1
                    break
                else:
                    routes_file_xmlroot.remove(trip_child)
                    break

    if has_ride == False:
        routes_file_xmlroot.remove(trip_child)
```

```python
for persontrip_child in list(persontrips_file_xmlroot):
    persontrip_in_routes = False
    for trip_child in list(routes_file_xmlroot):
        if trip_child.get("id") == persontrip_child.get("id"):
            persontrip_in_routes = True
            break
    
    if persontrip_in_routes == False:
        persontrips_file_xmlroot.remove(persontrip_child)
```

```python
persontrips_file_xmltree.write(persontrips_filepath_output)
routes_file_xmltree.write(routes_filepath_output)
```
