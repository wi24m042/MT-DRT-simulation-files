```python
import xml.etree.ElementTree as ET
import copy
```

```python
persontrips_filepath = "D:\\Dokumente\\sumo\\maloja\\09_daytrips\\persontrips_maloja_daytrips_summer.xml"
persontrips_filepath_output = "D:\\Dokumente\\sumo\\maloja\\09_daytrips\\persontrips_maloja_daytrips_summer_modded.xml"
routes_filepath = "D:\\Dokumente\\sumo\\maloja\\09_daytrips\\persontrips_maloja_daytrips_resolved_summer_bus.rou.xml"
routes_filepath_output = "D:\\Dokumente\\sumo\\maloja\\09_daytrips\\persontrips_maloja_daytrips_resolved_summer_bus_modded.rou.xml"
```

```python
demand_activities = {
    "taz_St_Moritz_Sink-taz_Cable_Car_Albigna": 61,
    "taz_St_Moritz_Sink-taz_Bregaglia_Summer": 61,
    "taz_St_Moritz_Sink-taz_Sils_Engadin_Summer": 61,
    "taz_St_Moritz_Sink-taz_Silvaplana_Summer": 61,
    "taz_St_Moritz_Sink-taz_St_Moritz_Summer": 61,
    "taz_St_Moritz_Sink-taz_ski_corviglia_piz_nair_st_moritz": 61,
    "taz_St_Moritz_Sink-taz_ski_corvatsch_sils_im_engadin": 61,

    "taz_St_Moritz_Sink-taz_St_Moritz": 107,
    "taz_St_Moritz_Sink-taz_Bregaglia": 107,
    "taz_St_Moritz_Sink-taz_Silvaplana": 107,
    "taz_St_Moritz_Sink-taz_Sils_Engadin": 107,
}

demand_additional = {
    "taz_Bregaglia-taz_St_Moritz": 27,
    "taz_Bregaglia-taz_Bregaglia": 27,
    "taz_Bregaglia-taz_Silvaplana": 27,
    "taz_Bregaglia-taz_Sils_Engadin": 27,

    "taz_Silvaplana-taz_St_Moritz": 27,
    "taz_Silvaplana-taz_Bregaglia": 27,
    "taz_Silvaplana-taz_Silvaplana": 27,
    "taz_Silvaplana-taz_Sils_Engadin": 27,

    "taz_Sils_Engadin-taz_St_Moritz": 27,
    "taz_Sils_Engadin-taz_Bregaglia": 27,
    "taz_Sils_Engadin-taz_Silvaplana": 27,
    "taz_Sils_Engadin-taz_Sils_Engadin": 27,

    "taz_St_Moritz-taz_St_Moritz": 27,
    "taz_St_Moritz-taz_Bregaglia": 27,
    "taz_St_Moritz-taz_Silvaplana": 27,
    "taz_St_Moritz-taz_Sils_Engadin": 27,
}
```

```python
demand_activities_from_sink = copy.deepcopy(demand_activities)
demand_activities_to_sink = copy.deepcopy(demand_activities)
```

```python
persontrips_file_xmltree = ET.parse(persontrips_filepath)
persontrips_file_xmlroot = persontrips_file_xmltree.getroot()
```

```python
routes_file_xmltree = ET.parse(routes_filepath)
routes_file_xmlroot = routes_file_xmltree.getroot()
```

```python
for trip_child in list(routes_file_xmlroot):
    has_ride = False
    for stage_child in list(trip_child):
        if stage_child.tag == "ride":
            if "intended" in stage_child.attrib:
                has_ride = True
                break

    if has_ride == True:
        for persontrip_child in list(persontrips_file_xmlroot):
            if trip_child.get("id") == persontrip_child.get("id"):
                demand_ft_key = str(persontrip_child[0].get("fromTaz")) + "-" + str(persontrip_child[0].get("toTaz"))
                demand_tf_key = str(persontrip_child[0].get("toTaz")) + "-" + str(persontrip_child[0].get("fromTaz"))
                
                if demand_ft_key in demand_activities_from_sink and demand_activities_from_sink[demand_ft_key] > 0:
                    demand_activities_from_sink[demand_ft_key] = demand_activities_from_sink[demand_ft_key] - 1
                    break
                elif demand_tf_key in demand_activities_to_sink and demand_activities_to_sink[demand_tf_key] > 0:
                    demand_activities_to_sink[demand_tf_key] = demand_activities_to_sink[demand_tf_key] - 1
                    break
                elif demand_ft_key in demand_additional and demand_additional[demand_ft_key] > 0:
                    demand_additional[demand_ft_key] = demand_additional[demand_ft_key] - 1
                    break
                else:
                    routes_file_xmlroot.remove(trip_child)
                    break

    if has_ride == False:
        routes_file_xmlroot.remove(trip_child)
```

```python
for persontrip_child in list(persontrips_file_xmlroot):
    persontrip_in_routes = False
    for trip_child in list(routes_file_xmlroot):
        if trip_child.get("id") == persontrip_child.get("id"):
            persontrip_in_routes = True
            break
    
    if persontrip_in_routes == False:
        persontrips_file_xmlroot.remove(persontrip_child)
```

```python
persontrips_file_xmltree.write(persontrips_filepath_output)
routes_file_xmltree.write(routes_filepath_output)
```
