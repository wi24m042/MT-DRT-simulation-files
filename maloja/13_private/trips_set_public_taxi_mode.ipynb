```python
import xml.etree.ElementTree as ET
```

```python
persontrips_summer_filepath = "D:\\Dokumente\\sumo\\maloja\\13_private\\persontrips_maloja_private_summer_modded.xml"
output_summer_filepath = "D:\\Dokumente\\sumo\\maloja\\13_private\\persontrips_maloja_private_summer_modded_public_taxi.xml"
persontrips_winter_filepath = "D:\\Dokumente\\sumo\\maloja\\13_private\\persontrips_maloja_private_winter_modded.xml"
output_winter_filepath = "D:\\Dokumente\\sumo\\maloja\\13_private\\persontrips_maloja_private_winter_modded_public_taxi.xml"
```

```python
persontrips_summer_file_xmltree = ET.parse(persontrips_summer_filepath)
persontrips_summer_file_xmlroot = persontrips_summer_file_xmltree.getroot()
```

```python
persontrips_winter_file_xmltree = ET.parse(persontrips_winter_filepath)
persontrips_winter_file_xmlroot = persontrips_winter_file_xmltree.getroot()
```

```python
for persontrip_summer_child in list(persontrips_summer_file_xmlroot):
    persontrip_summer_child[0].set("modes", "public taxi")
```

```python
for persontrip_winter_child in list(persontrips_winter_file_xmlroot):
    persontrip_winter_child[0].set("modes", "public taxi")
```

```python
persontrips_winter_file_xmltree.write(output_winter_filepath)
```

```python
persontrips_summer_file_xmltree.write(output_summer_filepath)
```
