```python
import xml.etree.ElementTree as ET
import copy
```

```python
persontrips_filepath = "D:\\Dokumente\\sumo\\maloja\\10_camping\\persontrips_maloja_camping_modded.xml"
output_filepath = "D:\\Dokumente\\sumo\\maloja\\10_camping\\persontrips_maloja_camping_modded_doubled.xml"
```

```python
persontrips_file_xmltree = ET.parse(persontrips_filepath)
persontrips_file_xmlroot = persontrips_file_xmltree.getroot()
```

```python
desitnations = ["taz_shopping_bregaglia_vicosoprano", "taz_shopping_silvaplana"]
```

```python
for person in list(persontrips_file_xmlroot):
    if float(str(person.get("depart"))) <= 50400 and (person[0].get("fromTaz") in desitnations or person[0].get("toTaz") in desitnations) and ("cgb" not in str(person.get("id")) or "cga" not in str(person.get("id")) or "cs" not in str(person.get("id")) or "ccga" not in str(person.get("id"))):
        person2 = copy.deepcopy(person)
        person2.set("id", "cga" + str(person2.get("id")))
        persontrips_file_xmlroot.insert(list(persontrips_file_xmlroot).index(person), person2)
        
        person.set("id", "cgb" + str(person.get("id")))
    elif (person[0].get("fromTaz") in desitnations or person[0].get("toTaz") in desitnations) and ("cgb" not in str(person.get("id")) or "cga" not in str(person.get("id")) or "cs" not in str(person.get("id")) or "ccga" not in str(person.get("id"))):
        person.set("id", "cs" + str(person.get("id")))
    elif "cgb" not in str(person.get("id")) or "cga" not in str(person.get("id")) or "cs" not in str(person.get("id")) or "ccga" not in str(person.get("id")):
        person.set("id", "c" + str(person.get("id")))
```

```python
persontrips_file_xmltree.write(output_filepath)
```
