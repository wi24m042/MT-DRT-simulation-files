```python
demand_shopping_from_camping_lt = copy.deepcopy(demand_shopping_lt)
demand_shopping_to_camping_lt = copy.deepcopy(demand_shopping_lt)
```

```python
demand_shopping_from_camping_st = copy.deepcopy(demand_shopping_st)
demand_shopping_to_camping_st = copy.deepcopy(demand_shopping_st)
```

```python
demand_summer_from_camping = copy.deepcopy(demand_summer)
demand_summer_to_camping = copy.deepcopy(demand_summer)
```

```python
persontrips_file_xmltree = ET.parse(persontrips_filepath)
persontrips_file_xmlroot = persontrips_file_xmltree.getroot()
```

```python
routes_file_xmltree = ET.parse(routes_filepath)
routes_file_xmlroot = routes_file_xmltree.getroot()
```

```python
for trip_child in list(routes_file_xmlroot):
    has_ride = False
    for stage_child in list(trip_child):
        if stage_child.tag == "ride":
            if "intended" in stage_child.attrib:
                has_ride = True
                break

    if has_ride == True:
        for persontrip_child in list(persontrips_file_xmlroot):
            if trip_child.get("id") == persontrip_child.get("id"):
                demand_ft_key = str(persontrip_child[0].get("fromTaz")) + "-" + str(persontrip_child[0].get("toTaz"))
                demand_tf_key = str(persontrip_child[0].get("toTaz")) + "-" + str(persontrip_child[0].get("fromTaz"))

                if 28800 <= float(str(trip_child.get("depart"))) <= 48600 and demand_ft_key in demand_shopping_from_camping_lt and demand_shopping_from_camping_lt[demand_ft_key] > 0:
                    demand_shopping_from_camping_lt[demand_ft_key] = demand_shopping_from_camping_lt[demand_ft_key] - 1
                    break
                elif 30600 <= float(str(trip_child.get("depart"))) <= 50400 and demand_tf_key in demand_shopping_to_camping_lt and demand_shopping_to_camping_lt[demand_tf_key] > 0:
                    demand_shopping_to_camping_lt[demand_tf_key] = demand_shopping_to_camping_lt[demand_tf_key] - 1
                    break
                elif 50400 <= float(str(trip_child.get("depart"))) <= 64800 and demand_ft_key in demand_shopping_from_camping_st and demand_shopping_from_camping_st[demand_ft_key] > 0:
                    demand_shopping_from_camping_st[demand_ft_key] = demand_shopping_from_camping_st[demand_ft_key] - 1
                    break
                elif 54000 <= float(str(trip_child.get("depart"))) <= 66600 and demand_tf_key in demand_shopping_to_camping_st and demand_shopping_to_camping_st[demand_tf_key] > 0:
                    demand_shopping_to_camping_st[demand_tf_key] = demand_shopping_to_camping_st[demand_tf_key] - 1
                    break
                elif demand_ft_key in demand_summer_from_camping and demand_summer_from_camping[demand_ft_key] > 0:
                    demand_summer_from_camping[demand_ft_key] = demand_summer_from_camping[demand_ft_key] - 1
                    break
                elif demand_tf_key in demand_summer_to_camping and demand_summer_to_camping[demand_tf_key] > 0:
                    demand_summer_to_camping[demand_tf_key] = demand_summer_to_camping[demand_tf_key] - 1
                    break
                else:
                    routes_file_xmlroot.remove(trip_child)
                    break

    if has_ride == False:
        routes_file_xmlroot.remove(trip_child)
```
