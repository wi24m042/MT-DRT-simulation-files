```python
import xml.etree.ElementTree as ET
```

```python
persontrips_filepath = "D:\\Dokumente\\sumo\\maloja\\10_camping\\persontrips_maloja_camping_modded_doubled.xml"
output_filepath = "D:\\Dokumente\\sumo\\maloja\\10_camping\\persontrips_maloja_camping_modded_doubled_public_taxi.xml"
```

```python
persontrips_file_xmltree = ET.parse(persontrips_filepath)
persontrips_file_xmlroot = persontrips_file_xmltree.getroot()
```

```python
for persontrip_child in list(persontrips_file_xmlroot):
    persontrip_child[0].set("modes", "public taxi")
```

```python
persontrips_file_xmltree.write(output_filepath)
```
