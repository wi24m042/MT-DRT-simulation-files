```python
import xml.etree.ElementTree as ET
import copy
```

```python
persontrips_filepath = "D:\\Dokumente\\sumo\\maloja\\11_hotel\\persontrips_maloja_hotel_winter.xml"
persontrips_filepath_output = "D:\\Dokumente\\sumo\\maloja\\11_hotel\\persontrips_maloja_hotel_winter_modded.xml"
routes_filepath = "D:\\Dokumente\\sumo\\maloja\\11_hotel\\persontrips_maloja_hotel_resolved_winter_bus.rou.xml"
routes_filepath_output = "D:\\Dokumente\\sumo\\maloja\\11_hotel\\persontrips_maloja_hotel_resolved_winter_bus_modded.rou.xml"
```

```python
demand_arriving_leaving = {
    "taz_St_Moritz_Sink-taz_Bregaglia": 23,
    "taz_St_Moritz_Sink-taz_Silvaplana": 37,
    "taz_St_Moritz_Sink-taz_Sils_Engadin": 79,
    "taz_St_Moritz_Sink-taz_St_Moritz": 303
}

demand_activities = {
    "taz_Bregaglia-taz_ski_aela": 135,
    
    "taz_Bregaglia-taz_St_Moritz": 12,
    "taz_Bregaglia-taz_Bregaglia": 12,
    "taz_Bregaglia-taz_Silvaplana": 12,
    "taz_Bregaglia-taz_Sils_Engadin": 12,
    "taz_Bregaglia-taz_St_Moritz_Sink": 12,

    "taz_Silvaplana-taz_ski_corvatsch_silvaplana": 279,

    "taz_Silvaplana-taz_St_Moritz": 24,
    "taz_Silvaplana-taz_Bregaglia": 24,
    "taz_Silvaplana-taz_Silvaplana": 24,
    "taz_Silvaplana-taz_Sils_Engadin": 24,
    "taz_Silvaplana-taz_St_Moritz_Sink": 24,
    
    "taz_Sils_Engadin-taz_ski_corvatsch_sils_im_engadin": 713,

    "taz_Sils_Engadin-taz_St_Moritz": 61,
    "taz_Sils_Engadin-taz_Bregaglia": 61,
    "taz_Sils_Engadin-taz_Silvaplana": 61,
    "taz_Sils_Engadin-taz_Sils_Engadin": 61,
    "taz_Sils_Engadin-taz_St_Moritz_Sink": 61,
    
    "taz_St_Moritz-taz_ski_corviglia_piz_nair_st_moritz": 2352,

    "taz_St_Moritz-taz_St_Moritz": 202,
    "taz_St_Moritz-taz_Bregaglia": 202,
    "taz_St_Moritz-taz_Silvaplana": 202,
    "taz_St_Moritz-taz_Sils_Engadin": 202,
    "taz_St_Moritz-taz_St_Moritz_Sink": 202,
}
```

```python
demand_hotel_arriving = copy.deepcopy(demand_arriving_leaving)
demand_hotel_leaving = copy.deepcopy(demand_arriving_leaving)
```

```python
demand_summer_from_hotel = copy.deepcopy(demand_activities)
demand_summer_to_hotel = copy.deepcopy(demand_activities)
```

```python
persontrips_file_xmltree = ET.parse(persontrips_filepath)
persontrips_file_xmlroot = persontrips_file_xmltree.getroot()
```

```python
routes_file_xmltree = ET.parse(routes_filepath)
routes_file_xmlroot = routes_file_xmltree.getroot()
```

```python
for trip_child in list(routes_file_xmlroot):
    has_ride = False
    for stage_child in list(trip_child):
        if stage_child.tag == "ride":
            if "intended" in stage_child.attrib:
                has_ride = True
                break

    if has_ride == True:
        for persontrip_child in list(persontrips_file_xmlroot):
            if trip_child.get("id") == persontrip_child.get("id"):
                demand_ft_key = str(persontrip_child[0].get("fromTaz")) + "-" + str(persontrip_child[0].get("toTaz"))
                demand_tf_key = str(persontrip_child[0].get("toTaz")) + "-" + str(persontrip_child[0].get("fromTaz"))

                if 36000 <= float(str(trip_child.get("depart"))) <= 43200 and demand_tf_key in demand_hotel_leaving and demand_hotel_leaving[demand_tf_key] > 0:
                    demand_hotel_leaving[demand_tf_key] = demand_hotel_leaving[demand_tf_key] - 1
                    break
                elif 43200 <= float(str(trip_child.get("depart"))) <= 50400 and demand_ft_key in demand_hotel_arriving and demand_hotel_arriving[demand_ft_key] > 0:
                    demand_hotel_arriving[demand_ft_key] = demand_hotel_arriving[demand_ft_key] - 1
                    break
                elif demand_ft_key in demand_summer_from_hotel and demand_summer_from_hotel[demand_ft_key] > 0:
                    demand_summer_from_hotel[demand_ft_key] = demand_summer_from_hotel[demand_ft_key] - 1
                    break
                elif demand_tf_key in demand_summer_to_hotel and demand_summer_to_hotel[demand_tf_key] > 0:
                    demand_summer_to_hotel[demand_tf_key] = demand_summer_to_hotel[demand_tf_key] - 1
                    break
                else:
                    routes_file_xmlroot.remove(trip_child)
                    break

    if has_ride == False:
        routes_file_xmlroot.remove(trip_child)
```

```python
for persontrip_child in list(persontrips_file_xmlroot):
    persontrip_in_routes = False
    for trip_child in list(routes_file_xmlroot):
        if trip_child.get("id") == persontrip_child.get("id"):
            persontrip_in_routes = True
            break
    
    if persontrip_in_routes == False:
        persontrips_file_xmlroot.remove(persontrip_child)
```
